 ## Q1. Remove missing data

Please remove the records that match ANY of the below two conditions in the csv file:

 * year is missing
 * fuel is 'none'

__Display the DataFrame, counting number of Null values in each column, and print the length of the data__ before and after removing the missing data.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Read the CSV file into a DataFrame
ds1 = 'data/car_sell.csv'
df1 = pd.read_csv(ds1)

# Display the DataFrame before cleaning the data
print("DataFrame before removing missing data:")
print(df1)

# Count the number of null values in each column before cleaning
null_counts_before = df1.isna().sum()
print("\nNumber of Null values in each column before cleaning:")
print(null_counts_before)

# Print the length (number of rows) of the DataFrame before cleaning
print("\nLength of the DataFrame before cleaning (number of rows):")
print(len(df1))

# Remove records where 'year' is missing OR 'fuel' is 'none'
df1_cleaned = df1[~(df1['year'].isna() | (df1['fuel'] == 'none'))]

# Display the DataFrame after cleaningthe data
print("\nDataFrame after removing missing data:")
print(df1_cleaned)

# Count the number of null values ineach column after cleaning the data
null_counts_after = df1_cleaned.isna().sum()
print("\nNumber of Null values in each column after cleaning:")
print(null_counts_after)

# Print the length (number of rows) of the DataFrame after cleaning the data
print("\nLength of the DataFrame after cleaning (number of rows):")
print(len(df1_cleaned))
```

## Q2. Descriptive statistics

With the cleaned data in Q1, please provide the data summarization as below:

* Q2.1 total number of unique name, unique seller_type, unique transmission, and unique owner
* Q2.2 descriptive statistics, including the total number, mean, std, min and max regarding all km_driven
* Q2.3 descriptive statistics, including mean, std, max, and min of the number of selling_price grouped by different seller_type
* Q2.4 descriptive statistics, including mean, std, max, min of the number of years that sold by each owner type


```python
import pandas as pd

# Assuming df1_cleaned is your DataFrame

# Calculate unique values
unique_names = df1_cleaned['name'].nunique()
unique_seller_types = df1_cleaned['seller_type'].nunique()
unique_transmissions = df1_cleaned['transmission'].nunique()
unique_owners = df1_cleaned['owner'].nunique()

# Descriptive statistics for 'km_driven'
km_driven_stats = df1_cleaned['km_driven'].describe()

# Descriptive statistics for 'selling_price' by 'seller_type'
seller_type_stats = df1_cleaned.groupby('seller_type')['selling_price'].agg(['mean', 'std', 'max', 'min'])

# Descriptive statistics for 'year' by 'owner'
owner_type_stats = df1_cleaned.groupby('owner')['year'].agg(['mean', 'std', 'max', 'min'])

# Print results
print(f"Total number of unique names: {unique_names}")
print(f"Total number of unique seller types: {unique_seller_types}")
print(f"Total number of unique transmissions: {unique_transmissions}")
print(f"Total number of unique owners: {unique_owners}")

print("\nDescriptive statistics for 'km_driven':")
print(km_driven_stats)

print("\nDescriptive statistics for 'selling_price' by 'seller_type':")
print(seller_type_stats)

print("\nDescriptive statistics for 'year' by 'owner':")
print(owner_type_stats)
```

```python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Selling Price Analysis by Different Categories', fontsize=20)

# Boxplot for 'fuel' vs 'selling_price'
axes[0, 0].boxplot([df1[df1['fuel'] == f]['selling_price'] for f in df1['fuel'].unique()],
                   labels=df1['fuel'].unique())
axes[0, 0].set_title('Selling Price by Fuel Type')
axes[0, 0].set_xlabel('Fuel Type')
axes[0, 0].set_ylabel('Selling Price')

# Boxplot for 'year' vs 'selling_price'
# Filter years to reduce clutter (e.g., every 2 years)
years = sorted(df1['year'].unique())
filtered_years = years[::2]  # Select every 2nd year

axes[0, 1].boxplot([df1[df1['year'] == y]['selling_price'] for y in filtered_years],
                   labels=filtered_years)
axes[0, 1].set_title('Selling Price by Year')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Selling Price')
axes[0, 1].grid(True, linestyle='--', alpha=0.7)


plt.setp(axes[0, 1].xaxis.get_majorticklabels(), rotation=45, ha="right")

# Boxplot for 'seller_type' vs 'selling_price'
axes[1, 0].boxplot([df1[df1['seller_type'] == s]['selling_price'] for s in df1['seller_type'].unique()],
                   labels=df1['seller_type'].unique())
axes[1, 0].set_title('Selling Price by Seller Type')
axes[1, 0].set_xlabel('Seller Type')
axes[1, 0].set_ylabel('Selling Price')

# Boxplot for 'transmission' vs 'selling_price'
axes[1, 1].boxplot([df1[df1['transmission'] == t]['selling_price'] for t in df1['transmission'].unique()],
                   labels=df1['transmission'].unique())
axes[1, 1].set_title('Selling Price by Transmission Type')
axes[1, 1].set_xlabel('Transmission Type')
axes[1, 1].set_ylabel('Selling Price')

# Boxplot for 'owner' vs 'selling_price'
axes[2, 0].boxplot([df1[df1['owner'] == o]['selling_price'] for o in df1['owner'].unique()],
                   labels=df1['owner'].unique())
axes[2, 0].set_title('Selling Price by Owner Type')
axes[2, 0].set_xlabel('Owner Type')
axes[2, 0].set_ylabel('Selling Price')


axes[2, 1].axis('off')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("1 Boxplot for 'fuel' vs 'selling_price'")
print("1.1 The x-axis is 'fuel' vs y-axis 'selling_price'")
print('1.2 Summary: Diesel cars generally have a higher median selling price than Petrol cars.')

print("\n\n")
print("2 Boxplot for 'year' vs 'selling_price'")
print("2.1 The x-axis is 'year' vs y-axis 'selling_price'")
print('2.2 Summary: The year 2018 and onwards the price of cars has in general gone up because thge median is comparitively higher in 2018 and onwards')
print("\n\n")
print("3 Boxplot for 'seller type' vs 'selling_price'")
print("3.1 The x-axis is 'seller type' vs y-axis 'selling_price'")
print('3.2 Summary: Trustmark  have a slight edge over the dealers as the median is higher but there are more outliers in dealers than Trustmark.')
print("\n\n")
print("4 Boxplot for 'Transmission type' vs 'selling_price'")
print("4.1 The x-axis is 'Transmission type' vs y-axis 'selling_price'")
print('4.2 Summary: Automatic is more expensive than manual as the median is higher and have more outliers in Automatic than manual.')

print("\n\n")
print("5 Boxplot for 'Owner type' vs 'selling_price'")
print("5.1 The x-axis is 'Owner type' vs y-axis 'selling_price'")
print('5.2 Summary: Test drive is the most expensive and the least expensive is fourth and above owner types')
# Explanation: Diesel engines are more fuel-efficient and durable, making them more valuable.

# Boxplot for 'year' vs 'selling_price'
```

## Q4. Detect and remove outliers

We define car sale records WITHOUT following ANY of the below three rules as outliers:

1. owner to be first, second and third owner only
2. Fuel to be Diesel or Petrol only
3. Seller_type to be dealer or individual only

Please remove the corresponding records in the csv file that is not qualified with the rules. After that, __print the length of the data__.

```python
valid_owners = ['First Owner', 'Second Owner', 'Third Owner']
valid_fuels = ['Diesel', 'Petrol']
valid_seller_types = ['Dealer', 'Individual']
print("The length of the unfiltered data is:", len(df1))
# Filter the DataFrame based on the defined rules
filtered_df = df1[
    (df1['owner'].isin(valid_owners)) &
    (df1['fuel'].isin(valid_fuels)) &
    (df1['seller_type'].isin(valid_seller_types))
]
# filtered_df.to_csv('data/car_sell.csv',index=False)
print(filtered_df)

print("\n\n")

# Print the length of the cleaned DataFrame
print("The length of the filtered data is:", len(filtered_df))
print("The count of outliers of the filtered data than the unfiltered is:", len(df1) - len(filtered_df))
```
